matplotlib 연습
========
간단한 그래프를 그려보자!

```python
import numpy as np
import matplotlib.pyplot as plot

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plot.plot(x, y1, label='sin')
plot.plot(x, y2, label='cos', linestyle='--')
plot.xlabel('x')
plot.ylabel('y')
plot.title('trigonometric function')
plot.legend()
plot.show()
```
