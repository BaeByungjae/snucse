```python
import numpy as np
import matplotlib.pyplot as plt

def draw(name, function):
    x = np.arange(-6, 6, 0.01)
    y = function(x)
    plt.plot(x, y)
    plt.title(name, fontsize=20)
    plt.show()

draw('step', lambda x: (x > 0).astype(int))
draw('sigmoid', lambda x: 1/(1 + np.exp(-x)))
```
