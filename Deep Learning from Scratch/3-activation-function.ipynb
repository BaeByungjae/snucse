```python
import numpy as np
import matplotlib.pyplot as plot

plot.figure(1, figsize=(9,7)).subplots_adjust(left=0.10, bottom=0.09, top=0.98, right=0.98)

# Plot activation functions
def function(label, function):
    x = np.arange(-2, 2, 0.01)
    y = function(x)
    plot.plot(x, y, label=label)
function('step', lambda x: (x > 0).astype(int))
function('sigmoid', lambda x: 1/(1 + np.exp(-x)))
function('relu', lambda x: np.maximum(0, x))
# function('polynomial', lambda x: 0.3*x*(x + 1.5)*(x - 1.5) + 1)

plot.legend()
plot.show()
```
